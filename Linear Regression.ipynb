```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
x_ = load_iris().data
y_ = load_iris().target
```

```python
x_df = pd.DataFrame(x_,columns=load_iris().feature_names)
y_df = pd.DataFrame(y_,columns=['target'])
raw_df = pd.concat([x_df,y_df],axis=1)
```

```python
raw_df
```

```python
filtered_df = raw_df[raw_df['target']==0]
```

```python
filtered_df
```

```python
r = filtered_df['sepal length (cm)']
c = filtered_df['sepal width (cm)']
```

```python
# assigning points to a and b so that the line is on the plot
a = np.array([5.7,3])
b = np.array([4.5,3])
```

```python
points = np.linspace(a,b,50)
```

```python
x_points = points[:,0]
y_points = points[:,1]
```

```python
plt.scatter(r,c)
plt.plot(x_points,y_points)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter plot of points')
plt.show()
```

```python
def plotl(w,b):
    x = np.linspace(4.2,5.9)
    y = w*x+b
    plt.plot(x,y)
    plt.scatter(r,c)
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.title('linear regression')
    plt.show()
```

```python
def calculate_mean_squared_error(r,c,m,y_c):
    error = ((m*r+y_c)-c)**2
    error = error.sum()
    return (error/len(r))/2
    
```

```python
def cost_d_m(r,c,m,y_c):
    temp = ((m*r+y_c)-c)*r
    temp = temp.sum()
    return temp/len(r)
```

```python
def cost_d_b(r,c,m,y_c):
    temp = ((m*r+y_c)-c)
    temp = temp.sum()
    return temp/len(r)
```

```python
# assign random values of m and y_c
m = 0.5
y_c = 0
print(calculate_mean_squared_error(r,c,m,y_c))
print(m,y_c)
plotl(m,y_c)
```

```python
#Run this cell and below repeatly
temp_m = m-0.01*cost_d_m(r,c,m,y_c)
temp_y_c = y_c-0.01*cost_d_b(r,c,m,y_c)
m = temp_m
y_c = temp_y_c
```

```python
print(calculate_mean_squared_error(r,c,m,y_c))
print(m,y_c)
plotl(m,y_c)
```
